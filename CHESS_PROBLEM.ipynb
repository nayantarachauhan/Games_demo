```python
def MinSteps(SourceX, SourceY, DestX, DestY): 

    # minimum number of steps 

    print(max(abs(SourceX - DestX), abs(SourceY - DestY))) 

    while ((SourceX != DestX) or (SourceY != DestY)): 

        # Go upward right 

        if ((SourceX < DestX) and (SourceY < DestY)): 

            print('UR-',end = "") 

            SourceX += 1
            SourceY += 1
            
        
        # Go upward left  
        
        if ((SourceX < DestX) and (SourceY > DestY)): 

            print('UL-',end = "") 

            SourceX += 1
            SourceY -= 1
            
        
        # Go downward left 
        
        if ((DestX < SourceX ) and (DestY < SourceY)): 

            print('DL-',end = "") 

            SourceX -= 1
            SourceY -= 1
            
         
        # Go downward right
        
        if ((SourceX > DestX) and (SourceY < DestY)): 

            print('DR-',end = "") 

            SourceX -= 1
            SourceY += 1
            

        # Go up 

        if ((SourceX < DestX) and (SourceY == DestY)): 

            print('U-',end = "") 

            SourceX += 1
            
            
        # Go down 

        if ((SourceX > DestX) and (SourceY == DestY)): 

            print('D-',end = "") 

            SourceX -= 1
            
               
        # Go left 

        if ((SourceX == DestX) and (SourceY > DestY)): 

            print('L-',end = "") 

            SourceY -= 1
            
               
        # Go right

        if ((SourceX == DestX) and (SourceY < DestY)): 

            print('R-',end = "") 

            SourceY += 1

```
